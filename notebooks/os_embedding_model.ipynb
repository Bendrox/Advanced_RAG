MTEB comprend 8 types de tches dembedding : 

	1.	Bitext Mining : trouver des phrases correspondantes dans deux langues diffrentes.

	2.	Classification : attribuer des catgories aux textes.

	3.	Clustering : regrouper des textes similaires.

	4.	Pair Classification : dterminer si deux textes sont similaires.

	5.	Reranking : ordonner une liste de textes en fonction de leur pertinence par rapport  une requte.

	6.	Retrieval : retrouver des documents pertinents pour une requte donne.

	7.	Semantic Textual Similarity (STS) : mesurer la similarit smantique entre deux textes.
